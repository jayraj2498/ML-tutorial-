# DBSCAN

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
centers = [[-1,-1],[2,2],[5,5]]
Xnw_db,ynw_db = make_blobs(n_samples = 500, centers=centers)
```

```python
set(ynw_db)
```

```python
colurset = np.array(['Red','blue','green','black'])
plt.scatter(Xnw_db[:,0],Xnw_db[:,1], c=colurset[ynw_db])
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps = 0.5, min_samples=6)
db.fit(Xnw_db)
dv_pred_db = db.labels_
```

```python
set(dv_pred_db)
```

```python
colurset = np.array(['Red','blue','green','black'])
plt.scatter(Xnw_db[:,0],Xnw_db[:,1], c=colurset[ynw_db])
```

```python
colurset = np.array(['Red','blue','green','black'])
plt.scatter(Xnw_db[:,0],Xnw_db[:,1], c=colurset[dv_pred_db])
```
