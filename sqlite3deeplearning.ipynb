```python
import sqlite3
```

```python
#step 1 creation of database(it can hold multiple table)
con = sqlite3.connect('first_database.db')
```

```python
#point out to the particular database where we want to create a table and insert data init
cur = con.cursor()
```

```python
#creation of table now
cur.execute("CREATE TABLE employee(empId, empName, empDesig)")
```
