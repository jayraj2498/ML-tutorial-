```python
import pandas as pd
data = pd.read_csv("cell_samples.csv")
```

```python
data.head(5)
```

```python
data.drop(['ID', 'BareNuc'],axis =1,inplace = True) 
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
X = data.drop(['Class'],axis = 1).values
y = data['Class'].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=50)
svc = SVC(kernel = 'rbf')
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
data = pd.read_csv('drug200.csv')
```

```python
data.head()
```

```python
set(data.Drug)
```

```python
X = data.drop(['Drug'],axis =1).values
y = data['Drug'].values
```

```python
from sklearn.preprocessing import LabelEncoder
leGender = LabelEncoder()
leGender.fit(['F','M'])
X[:,1] = leGender.transform(X[:,1])

from sklearn.preprocessing import LabelEncoder
leBp = LabelEncoder()
leBp.fit(['HIGH','LOW','NORMAL'])
X[:,2] = leBp.transform(X[:,2])


from sklearn.preprocessing import LabelEncoder
leBp = LabelEncoder()
leBp.fit(['HIGH','NORMAL'])
X[:,3] = leBp.transform(X[:,3])
```

```python
from sklearn.preprocessing import LabelEncoder
ledrug = LabelEncoder()
ledrug.fit(['drugA', 'drugB', 'drugC', 'drugX', 'drugY'])
y = ledrug.transform(y)
```

```python
X[:5]
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2, random_state=50)
tree =  DecisionTreeClassifier(criterion='entropy')
tree.fit(xtrain,ytrain)
ypred = tree.predict(xtest)
accuracy_score(ytest,ypred)
```
